Q1. Explain GET and POST methods.
Ans:HTTP (Hypertext Transfer Protocol) supports several methods, two of which are commonly used in web development: GET and POST.

GET method:

Used to request data from a specified resource.
Parameters are sent in the URL, visible to everyone.
Limited data can be sent, as it is appended to the URL.
POST method:

Used to submit data to be processed to a specified resource.
Parameters are sent in the request body, not visible in the URL.
Suitable for sending large amounts of data and sensitive information.
In summary, GET is used for retrieving data, while POST is used for submitting data.



Q2. Why is request used in Flask?
Ans:In Flask, the request object allows you to access incoming data from the client, such as form data, query parameters, and JSON data. It provides an easy way to handle the data sent by the client to the server. You can use request to extract information like form inputs, headers, and cookies from incoming HTTP requests.

Q3. Why is redirect() used in Flask?
Ans:In Flask, the redirect() function is used to redirect the client to a different location. This can be another route within the same Flask application or an external URL. It's commonly used after processing a form submission to redirect the user to a new page, preventing the user from resubmitting the form data if they refresh the page.

Q4. What are templates in Flask? Why is the render_template() function used?
Ans:Templates in Flask are used to separate the presentation layer (HTML code) from the application logic. They allow you to create dynamic HTML pages by embedding variables, control structures, and inheritance.

The render_template() function in Flask is used to render HTML templates. It takes the name of the template file as an argument, along with any variables you want to pass to the template. This function returns the HTML content generated by the template, which can then be sent to the client's browser.
